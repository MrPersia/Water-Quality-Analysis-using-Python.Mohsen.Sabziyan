```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")

data = pd.read_csv("water_potability.csv")
data.head()
```

```python
data.info()
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.countplot(x='Potability', data=data, palette='Set2')

plt.xlabel('Potability', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.title('Distribution of Unsafe and Safe Water', fontsize=16)

for p in plt.gca().patches:
    plt.gca().annotate('{:.0f}'.format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                        textcoords='offset points')

plt.show() #1 für trinkbar, 0 für nicht trinkbar
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="ph", hue="Potability", kde=True, multiple="stack", palette="husl", legend=True)
plt.title("Factors Affecting Water Quality: pH")
plt.xlabel("pH")
plt.ylabel("Count")
plt.legend(title="Potability", labels=["Potable", "Not Potable"], loc="upper right")
plt.tight_layout()
plt.show()
```

Die pH-Spalte stellt den pH-Wert des Wassers dar, der ein wichtiger Faktor bei der Beurteilung des Säure-Basen-Gleichgewichts des Wassers ist. Der pH-Wert des Trinkwassers sollte zwischen 6,5 und 8,5 liegen.

```python
import pandas as pd
from pycaret.classification import *

# Daten laden
data = pd.read_csv("water_potability.csv")

# Setup für Klassifikation mit einer bestimmten Session-ID
clf = setup(data, target="Potability", session_id=786)

# Modelle vergleichen und die besten Modelle auswählen
best_models = compare_models(sort='Accuracy', fold=5, n_select=5)

# Hyperparameter-Optimierung für die besten Modelle durchführen
tuned_models = [tune_model(model) for model in best_models]

# Feature Selection durchführen und die wichtigsten Merkmale anzeigen
for model in tuned_models:
    if hasattr(model, 'coef_') or hasattr(model, 'feature_importances_'):
        plot_model(model, plot='feature')

# Das beste Modell auswählen
best_model = automl(optimize='Accuracy')

# Modell auswerten
evaluate_model(best_model)
```
